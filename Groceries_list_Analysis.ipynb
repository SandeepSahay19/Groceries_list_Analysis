# About the Problem:
 
### **Here, I have been provided with the list of Groceries bought in two years time (01-01-2014 to 30-12-2015) from a local shop**. As a Data Analyst, I have to find Insights that will answer questions asked by the Sakeholders. Below are the questions provided :

# Questions:

# Time to find Insights

```python
import pandas as pd
df = pd.read_csv("Groceries.Sales_Analysis.csv")
df
```

```python
num_duplicates = df.duplicated(subset=['Member_number']).sum()
df
```

```python
Member_number_counts = df['Member_number'].value_counts().sort_values(ascending=False)

# Print the top 10 most common member number
print("Top 10 most common member number:")
print(Member_number_counts.head(10))
```

### Above is the top 10 list of Member_number with most shoppings 

```python
itemDescription_counts = df['itemDescription'].value_counts().sort_values(ascending=False)
print("Top 10 most common items bought:")
print(itemDescription_counts.head(10))
```

### Above is the list of top 15 items bought from the grocery shop 

```python
import matplotlib.pyplot as plt
plt.bar(itemDescription_counts.index[:10], itemDescription_counts[:10])
plt.xticks(rotation=90)
plt.xlabel('items')
plt.ylabel('Amount')
plt.title('Top 10 Most Bought items')
plt.show()
```

### Representation of Top 10 Most bought items

```python
Date_counts = df['Date'].value_counts().sort_values(ascending=False)
print("which Date was the busiest day:")
print(Date_counts.head(15))
```

### Above is the list of top 15 Date with most shoppings in a day or busiest Date.

```python
df['Date'] = pd.to_datetime(df['Date'], format='%d-%m-%Y')

# Extract the month information from the "date" column
df['month'] = df['Date'].dt.month

# Count the number of occurrences of each month and sort them in descending order
month_counts = df['month'].value_counts().sort_values(ascending=False)

# Print the most common month and its count
most_common_month = month_counts.index[0]
most_common_count = month_counts.iloc[0]
month_counts = df['month'].value_counts().sort_values()
month_counts.sort_values(ascending=False)
```

### Above is the list of items bought in every month

```python
import matplotlib.pyplot as plt
plt.bar(month_counts.index[:12], month_counts[:12])
plt.xticks(rotation=45)
plt.xlabel('months')
plt.ylabel('number of items')
plt.title('Number of items bought monthwise')
plt.show()
```
